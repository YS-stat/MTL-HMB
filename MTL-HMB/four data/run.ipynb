```python
import pandas as pd
import numpy as np
import os

# List of CSV file names to be processed
csv_files = [
    "HTL_four_data_gaussian_seeds=20.csv",
    "HTL_four_data_nongaussian_seeds=20.csv",
    "new_proposed_four_data_gaussian_seeds=20.csv",
    "new_proposed_four_data_nongaussian_seeds=20.csv",
    "SDL_four_data_gaussian_seeds=20.csv",
    "SDL_four_data_nongaussian_seeds=20.csv"
]

# Prepare an empty list to store summary results
result_rows = []

# Loop through each CSV file
for file in csv_files:
    df = pd.read_csv(file)

    # Automatically detect the three loss columns
    loss_cols = [col for col in df.columns if "loss" in col.lower()]

    # Skip file if less than 3 loss columns are found
    if len(loss_cols) != 4:
        print(f"Skipping {file}: less than 3 loss columns detected.")
        continue

    # Compute per-row (per-seed) statistics: mean, min, max of the three losses
    df["mean_loss"] = df[loss_cols].mean(axis=1)
    df["min_loss"] = df[loss_cols].min(axis=1)
    df["max_loss"] = df[loss_cols].max(axis=1)

    # Compute the overall mean and std of those per-seed statistics
    mean_of_mean = round(df["mean_loss"].mean(), 3)
    std_of_mean = round(df["mean_loss"].std(), 3)

    mean_of_min = round(df["min_loss"].mean(), 3)
    std_of_min = round(df["min_loss"].std(), 3)

    mean_of_max = round(df["max_loss"].mean(), 3)
    std_of_max = round(df["max_loss"].std(), 3)

    # Append the results to the summary list
    result_rows.append({
        "file": file,
        "mean(mean)": mean_of_mean,
        "std(mean)": std_of_mean,
        "mean(min)": mean_of_min,
        "std(min)": std_of_min,
        "mean(max)": mean_of_max,
        "std(max)": std_of_max
    })

# Convert the results to a DataFrame for visualization
results_df = pd.DataFrame(result_rows)
```
